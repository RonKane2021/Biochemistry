# This notebook is interactive widget for a normal Gaussian curve. The widgets let you change 3 parameters: center, amp, and sigma. 

```python
import ipywidgets as wd
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
%matplotlib nbagg
```

## Gaussian Curve equation to refresh

```python
def _1gaussian(amp1 ,cen1 ,sigma1):
    y = amp1*(1/(sigma1*(np.sqrt(2*np.pi))))*(np.exp((-1.0/2.0)*(((x-cen1)/sigma1)**2)))
    ax.clear()
    ax.plot(x,y)
    plt.show()
```

## Widget settings

```python
x = np.linspace(0,1000,1000)
fig, ax = plt.subplots(1, figsize = (9,4))
plt.suptitle("Gaussian Curve")

amp1 = wd.FloatSlider(min = 0.1, max = 100000, value =21861, description = "amp1")
cen1 = wd.FloatSlider(min = 0.1, max = 1000, value =509, description = "cen1")
sigma1 = wd.FloatSlider(min = 0.1, max = 1000, value =29.40, description = "sigma1")
wd.interactive(_1gaussian, amp1 = amp1,cen1 = cen1,sigma1 = sigma1)
```
