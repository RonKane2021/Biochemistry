### A small fun notebook which let's you play around with the Michaelis–Menten parameters Km and Vmax. By changing the values of both Km and Vmax, the graph will change in real-time to reflect those new parameters.

```python
import ipywidgets as wd
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
%matplotlib nbagg
```

### Define the MM equation

```python
def update_plot (Km, Vmax):
    ax.clear()
    y = (x*Vmax)/(x + Km)
    ax.plot(x,y)
    plt.xlabel('[Substrate]')
    plt.ylabel('Vo')
    plt.show()
```

### Using FloatSlider and interactive methods from ipwidgets to create the widgets.

```python
x = np.linspace(0,100,100)
fig, ax = plt.subplots(1, figsize = (9,4))
plt.suptitle("Enzyme Kinetics")

Km = wd.FloatSlider(min = 0.1, max = 100, value =1, description = "Km")
Vmax = wd.FloatSlider(min = 0.1, max = 100, value =1, description = "Vmax")
wd.interactive(update_plot, Km = Km, Vmax = Vmax)
```
