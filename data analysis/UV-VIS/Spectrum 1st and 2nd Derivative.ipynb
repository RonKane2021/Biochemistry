# Spectrum 1st and 2nd Derivative



### Here we will take a UV Spectrum from a csv file and plot the results and plot its 1st and 2nd derivative

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

### Load data from csv file into np.array.  delimeter is data seperated by "," . Check shape.

```python
a = np.loadtxt("gaus.csv", delimiter = ",")
a.shape
```

### Check array to see if data was correctly recorded.

```python
print (a[:,0])
print (a[:,1])
type (a)
```

### Plot the data to see the Spectrum

```python
x = (a[:,0])
y = (a[:,1])

plt.figure(figsize=(15,10))
plt.plot(x,y)
plt.show()
```

### Spectrum 1st Derivative. We will use np.diff method to differentiate along the y axis
https://numpy.org/doc/stable/reference/generated/numpy.diff.html

```python
n = 1
ydif_1 = np.diff(y)
xdif_1 = np.delete(a[:,0], n)

plt.figure(figsize=(15,10))
plt.plot(xdif_1,ydif_1)
plt.show()
```

## 2nd Derivative Spectrum. We will use np.diff method to differentiate along the y axis

```python
n=2
ydif_2 = np.diff(y, n =2)
xdif_1 = np.delete(a[:,0], n)

plt.figure(figsize=(15,10))
plt.plot(xdif_2,ydif_2)
plt.show()
```
