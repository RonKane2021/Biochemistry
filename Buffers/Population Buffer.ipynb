# Buffer Population Notebook

```python
import numpy as np
import matplotlib.pyplot as plt
```

## HEPES Example

```python
K1 = 1*10**-7.45 #pka of HEPES

def aHA(H, Ka):
    return (H)/(H + Ka)

def aH(H, Ka):
    return (Ka)/(H + Ka)

pH = np.linspace(1,14,100)

H = 10**-pH

y = aHA(H, K1)
y1 = aH(H, K1)

fig = plt.figure(figsize=(15,10))
plt.plot(pH, y, linestyle='--',label="HA")
plt.plot(pH, y1,linestyle='--', label= "A-")
plt.xlabel('pH')
plt.ylabel('Fraction')
plt.legend(loc='best')
plt.show()
```

## Glycine Example

```python
K1 = 1*10**-2.3 #carboxy terminal pKa
K2 = 1*10**-9.6 #amino terminal pKa

pH = np.linspace(0,14,1000)
H = 10**-pH

def aH2A(H, K1, K2):
    return (H**2)/(H**2 + (H*K1) + K1*K2)

def aHA(H, K1, K2):
    return (K1*H)/(H**2 + (H*K1) + K1*K2)

def aH(H, K1, K2):
    return (K1*K2)/(H**2 + (H*K1) + K1*K2)

y2 = aH2A(H, K1, K2)
y1 = aHA(H, K1, K2)
y = aH(H, K1, K2)
plt.figure(figsize=(15,10))
plt.plot(pH, y, linestyle='--', label = 'COO$^-$-Glycine-NH$_3$')
plt.plot(pH, y1, linestyle='--',  label = 'COO$^-$-Glycine-NH$_4$$^+$')
plt.plot(pH, y2, linestyle='--', label = 'COOH-Glycine-NH$_4$$^+$')
plt.xlabel('pH')
plt.ylabel('Fraction')
plt.legend(loc='best')
plt.show()
```

## Phosphoric Acid Example

```python
K1 = 7.1*10**-3
K2 = 6.3*10**-8
K3 = 2.1*10**-13
pH = np.linspace(0,14,1000)
H = 10**-pH

def aH3A(H, K1, K2, K3):
    return (H**3)/((H**3)+(H**2*K1)+(H*K1*K2)+(K1*K2*K3))

def aH2A(H, K1, K2, K3):
    return (H**2*K1)/((H**3)+(H**2*K1)+(H*K1*K2)+(K1*K2*K3))

def aHA(H, K1, K2, K3):
    return (H*K1*K2)/((H**3)+(H**2*K1)+(H*K1*K2)+(K1*K2*K3))
            
def aA(H, K1, K2, K3):
    return (K1*K2*K3)/((H**3)+(H**2*K1)+(H*K1*K2)+(K1*K2*K3))

y3 = aH3A(H, K1, K2, K3)
y2 = aH2A(H, K1, K2, K3)
y1 = aHA(H, K1, K2, K3)
y = aA(H, K1, K2, K3)

plt.figure(figsize=(15,10))
plt.plot(pH, y, linestyle='--', label = 'PO$_4$$^3$$^-$')
plt.plot(pH, y1, linestyle='--', label = 'HPO$_4$$^2$$^-$')
plt.plot(pH, y2, linestyle='--', label = 'H$_2$PO$_4$$^-$')
plt.plot(pH, y3, linestyle='--', label ='H$_3$PO$_4$')
plt.xlabel('pH')
plt.ylabel('Fraction')
plt.legend(loc='best')
plt.show()
```

# EDTA Example

```python
K1 = 1*10**-0.0
K2 = 1*10**-1.5
K3 = 1*10**-2
K4 = 1*10**-2.69
K5 = 1*10**-6.13
K6 = 1*10**-10.37

pH = np.linspace(0,14,1000)

H = 10**-pH

def aH6A(H, K1, K2, K3, K4, K5, K6):
    return (H**6)/((H**6)+(H**5*K1)+(H**4*K1*K2)+(H**3*K1*K2*K3)+(H**2*K1*K2*K3*K4)+(H*K1*K2*K3*K4*K5)+(K1*K2*K3*K4*K5*K6))

def aH5A(H, K1, K2, K3, K4, K5, K6):
    return (H**5*K1)/((H**6)+(H**5*K1)+(H**4*K1*K2)+(H**3*K1*K2*K3)+(H**2*K1*K2*K3*K4)+(H*K1*K2*K3*K4*K5)+(K1*K2*K3*K4*K5*K6))

def aH4A(H, K1, K2, K3, K4, K5, K6):
    return (H**4*K1*K2)/((H**6)+(H**5*K1)+(H**4*K1*K2)+(H**3*K1*K2*K3)+(H**2*K1*K2*K3*K4)+(H*K1*K2*K3*K4*K5)+(K1*K2*K3*K4*K5*K6))

def aH3A(H, K1, K2, K3, K4, K5, K6):
    return (H**3*K1*K2*K3)/((H**6)+(H**5*K1)+(H**4*K1*K2)+(H**3*K1*K2*K3)+(H**2*K1*K2*K3*K4)+(H*K1*K2*K3*K4*K5)+(K1*K2*K3*K4*K5*K6))

def aH2A(H, K1, K2, K3, K4, K5, K6):
    return (H**2*K1*K2*K3*K4)/((H**6)+(H**5*K1)+(H**4*K1*K2)+(H**3*K1*K2*K3)+(H**2*K1*K2*K3*K4)+(H*K1*K2*K3*K4*K5)+(K1*K2*K3*K4*K5*K6))

def aHA(H, K1, K2, K3, K4, K5, K6):
    return (H*K1*K2*K3*K4*K5)/((H**6)+(H**5*K1)+(H**4*K1*K2)+(H**3*K1*K2*K3)+(H**2*K1*K2*K3*K4)+(H*K1*K2*K3*K4*K5)+(K1*K2*K3*K4*K5*K6))

def aA(H, K1, K2, K3, K4, K5, K6):
    return (K1*K2*K3*K4*K5*K6)/((H**6)+(H**5*K1)+(H**4*K1*K2)+(H**3*K1*K2*K3)+(H**2*K1*K2*K3*K4)+(H*K1*K2*K3*K4*K5)+(K1*K2*K3*K4*K5*K6))

y6 = aH6A(H, K1, K2, K3, K4, K5, K6)
y5 = aH5A(H, K1, K2, K3, K4, K5, K6)
y5 = aH5A(H, K1, K2, K3, K4, K5, K6)   
y4 = aH4A(H, K1, K2, K3, K4, K5, K6)
y3 = aH3A(H, K1, K2, K3, K4, K5, K6)
y2 = aH2A(H, K1, K2, K3, K4, K5, K6)
y1 = aHA(H, K1, K2, K3, K4, K5, K6)
y = aA(H, K1, K2, K3, K4, K5, K6)

plt.figure(figsize=(15,10))
plt.plot(pH, y,  linestyle='--', label = 'Y$^4$$^-$')
plt.plot(pH, y1,  linestyle='--', label = 'HY$^3$$^-$')
plt.plot(pH, y2,  linestyle='--', label = 'H$_2$Y$^2$$^-$')
plt.plot(pH, y3,  linestyle='--', label = 'H$_3$Y$^-$')
plt.plot(pH, y4,  linestyle='--', label = 'H$_4$Y')
plt.plot(pH, y5,  linestyle='--', label = 'H$_5$Y$^+$')
plt.plot(pH, y6,  linestyle='--', label = 'H$_6$Y$^2$$^+$')
plt.xlabel('pH')
plt.ylabel('Fraction')
plt.legend(loc='best')
plt.show()
```
